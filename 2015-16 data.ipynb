```python
import pandas as pd
full_data=pd.read_csv("15-16_missing_data.csv")
data=full_data[["No of Vehicles", "No of Accidents", "Fuel Liters", "Passenger Kms.", "Vehicle Kms. (in lakh)", "No. of Labour", "Passenger Carried"]]
```

```python
#missing values in each column
data.isnull().sum() 
```

```python
#total missing values
data.isnull().sum().sum()
```

```python
data
```

```python
#defining imputer
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=4, weights='uniform', metric='nan_euclidean')
```

```python
#fitting imouter on dataset
imputer.fit(data)
```

```python
#transforming the data
data_trans=imputer.transform(data)
data_trans
```
